# Gidler BAA - Agent Player (Kaggle Edition)
This notebook runs the Branching-Attention Agent (BAA) on the Humanoid-v5 environment in headless mode.
It is designed to run for extended periods (Infinite Life Mode) and save the learned model weights for download.

```python
# 1. Setup Framework
import os
import shutil
import sys

INPUT_FOLDER = '/kaggle/input/gidler-baa-framework-2026-v1'
WORKING_DIR = '/kaggle/working/gidler-baa-framework'

print("🚀 Setting up Gidler Framework...")

if not os.path.exists(WORKING_DIR):
    os.makedirs(WORKING_DIR)

if os.path.isdir(INPUT_FOLDER):
    print(f"📂 Copying from {INPUT_FOLDER}...")
    for item in os.listdir(INPUT_FOLDER):
        s = os.path.join(INPUT_FOLDER, item)
        d = os.path.join(WORKING_DIR, item)
        if os.path.isdir(s):
            if os.path.exists(d): shutil.rmtree(d)
            shutil.copytree(s, d)
        else:
            shutil.copy2(s, d)
    print(f"✅ Framework installed to {WORKING_DIR}")
else:
    print(f"❌ ERROR: {INPUT_FOLDER} not found! Check your dataset mapping.")
```
